```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data Loading
```

```python
# To display all columns
pd.set_option('display.max_columns', None)

# Load the dataset
data = pd.read_csv("SuperStore.csv")

# Display first 10 rows
print("First 10 rows of the dataset:")
display(data.head(10))

# Show shape, column names, and basic info
print("\nShape of the dataset:", data.shape)
print("\nColumn names:\n", data.columns.tolist())
print("\nBasic Info:")
print(data.info())

# Display basic statistical summary
print("\nSummary Statistics:")
display(data.describe())
```

```python
# Data Cleaning
```

```python
# Check for missing values
print("Missing Values in Each Column:\n")
print(data.isnull().sum())

# Handle missing values properly (without inplace warning)
for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = data[col].fillna(data[col].mode()[0])
    else:
        data[col] = data[col].fillna(data[col].mean())

print("\nAfter handling missing values:\n")
print(data.isnull().sum())

# Check data types
print("\nData Types of Each Column:\n")
print(data.dtypes)

# Convert data types if necessary (example: Order Date & Ship Date to datetime)
if 'Order Date' in data.columns:
    data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')

if 'Ship Date' in data.columns:
    data['Ship Date'] = pd.to_datetime(data['Ship Date'], errors='coerce')

print("\nData types after conversion:\n")
print(data.dtypes)

# Check for duplicate rows
duplicates = data.duplicated().sum()
print(f"\nNumber of Duplicate Rows: {duplicates}")

# Drop duplicates if any
if duplicates > 0:
    data = data.drop_duplicates().reset_index(drop=True)
    print("Duplicates removed successfully.")
else:
    print("No duplicates found.")
```

```python
# Data Visualization
```

```python
# Bar Plot
```

```python
# Set general plot style
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(10, 5))

# 1️⃣ Sales by Category
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Sales', data=data, estimator='sum', palette='pastel')
plt.title('Total Sales by Category')
plt.show()

# 2️⃣ Sales by Sub-Category
plt.figure(figsize=(10,6))
sns.barplot(x='Sub-Category', y='Sales', data=data, estimator='sum', palette='Set2')
plt.title('Total Sales by Sub-Category')
plt.xticks(rotation=45)
plt.show()

# 3️⃣ Sales by Segment
plt.figure(figsize=(8,5))
sns.barplot(x='Segment', y='Sales', data=data, estimator='sum', palette='muted')
plt.title('Total Sales by Segment')
plt.show()

# 4️⃣ Sales by Region
plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Sales', data=data, estimator='sum', palette='cool')
plt.title('Total Sales by Region')
plt.show()

# 5️⃣ Top 10 States by Sales
top_states = data.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_states.values, y=top_states.index, palette='viridis')
plt.title('Top 10 States by Total Sales')
plt.xlabel('Sales')
plt.ylabel('State')
plt.show()
```

```python
# Scatter Plot
```

```python
# 1️⃣ Sales vs Profit
plt.figure(figsize=(8,5))
sns.scatterplot(x='Sales', y='Profit', data=data, hue='Category', palette='tab10', alpha=0.8)
plt.title('Sales vs Profit (by Category)')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()

# 2️⃣ Discount vs Profit
plt.figure(figsize=(8,5))
sns.scatterplot(x='Discount', y='Profit', data=data, hue='Segment', palette='Set1', alpha=0.8)
plt.title('Discount vs Profit (by Segment)')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

```python
# Histogram 
```

```python
# --- Histogram for numeric variables ---
data[['Sales', 'Profit', 'Quantity', 'Discount']].hist(bins=20, figsize=(10,6))
plt.suptitle("Distribution of Numeric Features", fontsize=14)
plt.show()
```

```python
# heapmap
```

```python
# Select only numeric columns
numeric_data = data[['Sales', 'Profit', 'Quantity', 'Discount']]

# Compute correlation matrix
corr = numeric_data.corr()

# Plot heatmap
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
plt.title('Correlation Heatmap of Numeric Variables')
plt.show()
```

```python
# Outliers detection
```

```python
# Columns to check for outliers
cols = ['Sales', 'Profit', 'Discount']

# Create boxplots for each column
for col in cols:
    plt.figure(figsize=(7,4))
    sns.boxplot(data=data, x=col, color='lightcoral')
    plt.title(f'Boxplot of {col} (Outlier Detection)')
    plt.xlabel(col)
    plt.show()
```
